```python
DB_HOST = "localhost"
DB_PORT = 3306
DB_USER = "root"
DB_PASSWORD = "aiproject"
DB_NAME = "chatbot"
DB_CHARSET = "utf8"
```

```python
import pymysql

db = None
try:
    db = pymysql.connect(
        host= DB_HOST,
        user=DB_USER,
        port=DB_PORT,
        passwd=DB_PASSWORD,
        db=DB_NAME,
        charset=DB_CHARSET
    )

    # 테이블 생성 sql 정의
    sql = '''
    CREATE TABLE IF NOT EXISTS `chatbot` (
    `id` INT UNSIGNED NOT NULL AUTO_INCREMENT,
    `intent` VARCHAR(45) NULL,
    `ner` VARCHAR(1024) NULL,
    `query` TEXT NULL,
    `answer` TEXT NOT NULL,
    `answer_image` VARCHAR(2048) NULL,
    PRIMARY KEY (`id`))
    ENGINE = InnoDB DEFAULT CHARSET=utf8
    '''

    # 테이블 생성
    with db.cursor() as cursor:
        cursor.execute(sql)

except Exception as e:
    print(e)

finally:
    if db is not None:
        db.close()

```

```python
import pymysql
import openpyxl


# 학습 데이터 초기화
def all_clear_train_data(db):
    # 기존 학습 데이터 삭제
    sql = '''
            delete from chatbot
        '''
    with db.cursor() as cursor:
        cursor.execute(sql)

    # auto increment 초기화
    sql = '''
    ALTER TABLE chatbot AUTO_INCREMENT=1
    '''
    with db.cursor() as cursor:
        cursor.execute(sql)


# db에 데이터 저장
def insert_data(db, xls_row):
    intent, ner, query, answer, answer_img_url = xls_row

    sql = '''
        INSERT chatbot(intent, ner, query, answer, answer_image) 
        values('%s', '%s', '%s', '%s', '%s'
        )
    ''' % (intent.value, ner.value, query.value, answer.value, answer_img_url.value)

    # 엑셀에서 불러온 cell에 데이터가 없는 경우, null 로 치환
    sql = sql.replace("'None'", "null")

    with db.cursor() as cursor:
        cursor.execute(sql)
        print('{} 저장'.format(query.value))
        db.commit()


train_file = "raw_data.xlsx" # 이 데이터 우리껄로 바꾸기!!!!!!!!!!!!!
db = None
try:
    db = pymysql.connect(
        host= DB_HOST,
        user=DB_USER,
        port=DB_PORT,
        passwd=DB_PASSWORD,
        db=DB_NAME,
        charset=DB_CHARSET
    )

    # 기존 학습 데이터 초기화
    all_clear_train_data(db)

    # 학습 엑셀 파일 불러오기
    wb = openpyxl.load_workbook(train_file)
    sheet = wb['Sheet1']
    for row in sheet.iter_rows(min_row=2): # 해더는 불러오지 않음
        # 데이터 저장
        insert_data(db, row)

    wb.close()

except Exception as e:
    print(e)

finally:
    if db is not None:
        db.close()

```
